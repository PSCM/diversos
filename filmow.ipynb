# Bibliotecas

```python
from urllib.parse import urljoin
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

# Função

```python
# Função para verificar e coletar o total de páginas associadas ao usuário 
def coletarFilmes(usuario):
  headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/61.0.3163.100 Safari/537.36'}

  paginas  = []
  filmes = []

  with requests.Session() as session:
    pagina_numero = 1
    url = "https://filmow.com/usuario/"+usuario+"/filmes/ja-vi/?pagina="
    
    while True:
        response = session.get(url, headers=headers)
        soup = BeautifulSoup(response.content, 'html.parser')
        paginas.append(url)

        # Verifica se há próxima página; quebra, caso não
        next_link = soup.find("a", id="next-page")
        if next_link is None:
            break

        url = urljoin(url, next_link["href"])
        pagina_numero += 1
    
    # Coleta os títulos dos filmes assistidos pelo usuário
    for pagina in paginas:
      response = session.get(pagina, headers=headers)
      soup = BeautifulSoup(response.content, 'html.parser')
      for div in soup.select("div.content-list"):
       for span in div.find_all("span", class_="title"):
        filmes.append(span.text)

  df = pd.DataFrame(filmes, columns=['filmes'])

  return(df)
```

```python
# Criar um data frame com os títulos assistidos e listados pelo usuário
df = coletarFilmes("aninhapaulaana") # Usuário com perfil público, listado na primeira página da plataforma
df.head(25)
```
